```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","C","C++","Java","js"]
y=[80,45,75,70,80]
z=[70,60,70,60,65]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Programming Language",fontsize=15)
plt.ylabel("Popularity Percentage",fontsize=15)
plt.title("Yearwise",fontsize=25)

plt.bar(p,y,width,color="g",label="Year 2025")
plt.bar(p1,z,width,color="b",label="Year 2024")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[4,5,1,9,7,3]
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g")
plt.title("Fill_Between Plot",fontsize=25)
plt.xlabel("X-Axis",fontsize=12)
plt.ylabel("Y-Axis",fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6])
y=np.array([4,5,1,9,7,3])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=5))
plt.title("Fill_Between Plot",fontsize=25)
plt.xlabel("X-Axis",fontsize=12)
plt.ylabel("Y-Axis",fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6])
y=np.array([4,5,1,9,7,3])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=5),alpha=0.6)
plt.title("Fill_Between Plot",fontsize=25)
plt.xlabel("X-Axis",fontsize=12)
plt.ylabel("Y-Axis",fontsize=12)
plt.show()
```
