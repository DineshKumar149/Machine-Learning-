PIVOT()

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],
                 "st_name":["a","b","c","a","b","c"],"eng":[10,12,13,14,12,19],"math":[13,17,39,43,14,18]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

```python
var.pivot_table(index="st_name",columns="days",
         aggfunc="mean")
```

```python
var.pivot_table(index="st_name",columns="days",
         aggfunc="sum")
```

```python
var.pivot_table(index="st_name",columns="days",
                aggfunc="sum",margins="True")
```
